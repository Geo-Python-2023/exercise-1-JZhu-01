The output should contain mean, max, and min Celsius temperatures for each day (for example, one mean temperature value for the 1st of May and so on).

````python
# YOUR CODE HERE
import pandas as pd
# Read data from the CSV file
kumpula = pd.read_csv('Kumpula_temps_May_Aug_2017.csv')

# Delete NaN columns
kumpula.dropna(axis=1, inplace=True)

# Delete the TEMP column
kumpula.drop(['TEMP'], axis=1, inplace=True)
````

````python
kumpula.head()
````

````python
# Convert to datetime format
kumpula['YR--MODAHRMN'] = pd.to_datetime(kumpula['YR--MODAHRMN'], format='%Y%m%d%H%M')

# Create a new column for the date
kumpula['YR--MODAHRMN'] = kumpula['YR--MODAHRMN'].dt.date

# Rename the column
kumpula.rename(columns={'YR--MODAHRMN':'DATE'}, inplace=True)
````

````python
kumpula.head()
````

````python
# Group the dataframe by DATE and perform calculations
kumpula_daily = kumpula.groupby('DATE', as_index=False)['Celsius'].agg(['mean', 'max', 'min'])
kumpula_daily.head()
````

````python
# Round the data to one decimal place
kumpula_daily['mean'] = kumpula_daily['mean'].round(1)
kumpula_daily.head()
````

````python
# Add a new column for station code
station_code = pd.Series([29980] * len(kumpula_daily), index=kumpula_daily.index, name='USAF')

# Move the USAF column to the first position
kumpula_daily.insert(0, 'USAF', station_code)
kumpula_daily.head()
````
